```python
df["date"] = pd.to_datetime(df["date"])
df = df.groupby(pd.Grouper(key='date', freq='1M')).mean()
df.index = df.index.strftime('%B')
df.plot()
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
